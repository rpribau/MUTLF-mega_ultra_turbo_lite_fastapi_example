```python
import pandas as pd
import requests

url = "http://127.0.0.1:8000/predict"

# Mapeo de los resultados de regreso a las etiquetas originales que usamos en el entrenamiento
labels_map = {0: "Negative", 1: "Neutral", 2: "Positive"}

# Frases ahora orientadas específicamente a Sentimiento Financiero
financial_sentences = [
    "Operating profit rose to EUR 13.1 mn from EUR 8.7 mn in the corresponding period in 2007 representing 7.7 % of net sales.",
    "Sales in Finland decreased by 10.5 % in January, while sales outside Finland dropped by 17 %.",
    "The board of directors is pleased to announce a dividend of 0.50 USD per share.",
    "Tesla stocks tumbled heavily after the unexpected market crash.",
    "The company expects net sales to remain at the same level as last year."
]

results = []

for sentence in financial_sentences:
    response = requests.post(url, json={"text": sentence})
    
    if response.status_code == 200:
        data = response.json()
        label_text = labels_map.get(data["prediction"], "Unknown")
        prob_percentage = f"{data['probability']:.2%}"
        
        results.append({
            "Financial Statement": sentence,
            "Sentiment": label_text,
            "Confidence": prob_percentage
        })
    else:
        print(f"Error ({response.status_code}) on statement: '{sentence}'")

# Visualizar los resultados en una tabla atractiva con Pandas
df_results = pd.DataFrame(results)
df_results
```
